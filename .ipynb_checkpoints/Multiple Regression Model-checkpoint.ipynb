## Training the model

### Building a linear Model 

<h3> APPROACH USED :</h3>
<p>We will use a mixed approach to build the model.
Here we are using RFE approach for feature selection and then we will use the statsmodel approach for building the model </p>

```python
# Running RFE with the output number of the variable equal to 12
lm = LinearRegression()
lm.fit(x_train, y_train)

rfe = RFE(lm, 12)             # running RFE
rfe = rfe.fit(x_train, y_train)
```

```python
list(zip(x_train.columns,rfe.support_,rfe.ranking_))
```

```python
col = x_train.columns[rfe.support_]
col
```

```python
x_train.columns[~rfe.support_]
```

```python
# Creating X_test dataframe with RFE selected variables
X_train_rfe = x_train[col]
X_train_rfe = sm.add_constant(X_train_rfe)
```

```python
X_train_rfe = X_train_rfe.apply(pd.to_numeric, errors='coerce')
```

```python
lm1 = sm.OLS(y_train,X_train_rfe).fit()   # Running the linear model
```

```python
#Let's see the summary of our linear model
print(lm1.summary())
```

<h3> Here we see that the p-value for all the variables is < 0.05 . Hence, we keep all the columns and proceed with the model.</h3>

### Check for VIF to eliminate multi collinearity in the model

<p>Variance Inflation Factor or VIF, gives a basic quantitative idea about how much the feature variables are correlated with each other. It is an extremely important parameter to test our linear model.</p>



```python
X_train_new = X_train_rfe.drop(['const'], axis=1)
```

```python
vif = pd.DataFrame()
X = X_train_new
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

##### 'Hum' has high VIF of 23.20 and so needs to be removed from the model, acceptable range is 5

```python
X_train_new1 = X_train_new.drop(["hum"], axis = 1)
```

```python
X_train_lm = sm.add_constant(X_train_new1)
lm1 = sm.OLS(y_train,X_train_lm).fit() 
print(lm1.summary())
```

```python
X_train_new2 = X_train_lm.drop(['const'], axis=1)
```

```python
vif = pd.DataFrame()
X = X_train_new2
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

#### The VIF value of temp is nearly equal to 5. Hence, we are not dropping this feature.
